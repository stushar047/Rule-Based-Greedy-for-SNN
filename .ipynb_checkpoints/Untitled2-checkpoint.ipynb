```python
import json
import random
import os
import numpy as np
import pandas as pd
import tensorflow as tf


# -------------------------------
# Config
# -------------------------------
with open("C:/Users/somas/Paper_exp/Enc_Dec_v7/config.json", "r") as config_file:
    Config = json.load(config_file)

Config["file"] = "C:/Users/somas/Paper_exp/Enc_Dec_v7/Mem_data/3T1R_New.xlsx"
Config["Res_Column"] = "Read Res."
Config["Power_Column"] = "Read_Power (µW)"


# -------------------------------
# Helpers
# -------------------------------
def set_seed(seed_value):
    os.environ["PYTHONHASHSEED"] = str(seed_value)
    random.seed(seed_value)
    np.random.seed(seed_value)
    tf.random.set_seed(seed_value)


def load_model_weights_from_csv(csv_path):
    """
    Load weight matrix from CSV.
    Expected shape: (input_dim, output_dim)
    """
    W = pd.read_csv(csv_path, header=None).values.astype(np.float32)
    return W


def generate_random_binary_input(num_samples, input_dim):
    """
    Generate random binary spike input.

    Output shape: (num_samples, input_dim, f_sample)
    """
    return np.random.randint(
        0, 2, size=(num_samples, input_dim)
    ).astype(np.float32)


def weight_to_resistance(W, R_max, R_min):
    """
    Map weight matrix to resistance values using linear conductance scaling.

    W shape: (input_dim, output_dim)
    Returns: resistance matrix of shape (input_dim, output_dim)
    """
    G_H = 1.0 / (R_min * 1e3)
    G_L = 1.0 / (R_max * 1e3)

    w_min = np.min(W)
    w_max = np.max(W)

    alpha = (G_H - G_L) / (w_max + abs(w_min))
    G = alpha * (W + abs(w_min)) + G_L

    R = ((G ** -1) * 1e-3).astype(int)
    return R


def table_creation(config):
    """
    Read lookup table from Excel and create TensorFlow hash table.
    Uses resistance -> power mapping.
    """
    df = pd.read_excel(config["file"])

    R_table = df[config["Res_Column"]].values.astype(int)
    P_table = df[config["Power_Column"]].values.astype(np.float32)

    R_min, R_max = R_table[0], R_table[-1]

    keys_tensor = tf.constant(R_table)
    vals_tensor = tf.constant(P_table)

    init = tf.lookup.KeyValueTensorInitializer(keys_tensor, vals_tensor)
    table = tf.lookup.StaticHashTable(init, default_value=-1.0)

    return table, R_min, R_max


def compute_mean_power(X_binary, power_matrix):
    """
    Compute output power over time.

    X_binary shape: (num_samples, input_dim, T)
    power_matrix shape: (input_dim, output_dim)

    Returns:
      mean_power: scalar
      Y_power: shape (num_samples, output_dim, T)
    """
    num_samples, input_dim = X_binary.shape
    output_dim = power_matrix.shape[1]
    
    Y_power = np.dot(X_binary, power_matrix)

#     for t in range(T):
#         # (num_samples, input_dim) @ (input_dim, output_dim)
#         Y_power[:, :, t] = X_binary[:, :, t] @ power_matrix

    mean_power = np.mean(np.sum(Y_power,axis=1))
    return mean_power, Y_power


# -------------------------------
# Main
# -------------------------------
model_csv_path = "C:/Users/somas/Downloads/fc2_weight.csv"   # change this
num_samples = 100
seed_value = 1
f_sample=1

set_seed(seed_value)

# Load memristor lookup table
table, R_min, R_max = table_creation(Config)

# Load model weights from CSV
W = load_model_weights_from_csv(model_csv_path).T
input_dim, output_dim = W.shape

print(f"Loaded weight matrix shape: {W.shape}")

# Generate random binary input
X_binary = generate_random_binary_input(
    num_samples=num_samples,
    input_dim=input_dim
)

print(f"Random binary input shape: {X_binary.shape}")

# Map weights to resistance
R = weight_to_resistance(W, R_max, R_min)

# Lookup power matrix from resistance matrix
power_matrix = table.lookup(tf.constant(R)).numpy()

print(f"Power matrix shape: {power_matrix.shape}")

# Compute mean power
mean_power, Y_power = compute_mean_power(X_binary, power_matrix)

print(f"Mean Power = {mean_power:.6e} µW")
```
