```python
import json
import numpy as np
import matplotlib.pyplot as plt

with open("raw_metrics_wine_v2.json") as file:
    
    raw_wine = json.load(file)
    
with open("raw_metrics_iris_v2.json") as file:
    
    raw_iris = json.load(file)
    
with open("raw_metrics_bc_v2.json") as file:
    
    raw_bc = json.load(file)
```

```python
# Define datasets and input files
datasets = {
    "wine": "raw_metrics_wine_v2.json",
    "iris": "raw_metrics_iris_v2.json",
    "bc": "raw_metrics_bc_v2.json"
}

f_samples = [8, 16, 32, 64]

# Initialize summaries
summary_wine = {}
summary_iris = {}
summary_bc = {}

# Loop through each dataset
for name, file_path in datasets.items():
    with open(file_path, "r") as file:
        metrics = json.load(file)

    summary = {}
    for f_sample in f_samples:
        f_sample_str = str(f_sample)
        summary[f_sample_str] = {}

        for key, values in metrics[f_sample_str].items():
            # Keep confusion matrix entries as-is
            if key.startswith("confusion"):
                pass
            else:
                arr = np.array(values, dtype=float)
                summary[f_sample_str][key] = {
                    "mean": float(np.mean(arr)),
                    "std": float(np.std(arr))
                }

    # Assign to correct variable
    if name == "wine":
        summary_wine = summary
    elif name == "iris":
        summary_iris = summary
    elif name == "bc":
        summary_bc = summary

plt.rcParams.update({
    'font.weight': 'bold',
    'axes.labelweight': 'bold',
    'axes.titleweight': 'bold',
    'xtick.labelsize': 26,
    'ytick.labelsize': 26,
    'xtick.major.width': 2,
    'ytick.major.width': 2,
})

def plot_metric_subplot(ax, summary, f_samples, metric_keys, ylabel, title,
                        offset_width=0.05, font_label=26, font_tick=26,
                        line_width=2, marker_size=6, cap_size=4):
    """
    Plot mean ± std for given metrics on a provided Axes (subplot).
    """
    num_keys = len(metric_keys)
    cmap = plt.cm.tab10(np.linspace(0, 1, num_keys))
    offsets = np.linspace(-offset_width, offset_width, num_keys)

    for i, key in enumerate(metric_keys):
        means, stds, valid_fs = [], [], []
        for f in f_samples:
            if key in summary[str(f)]:
                means.append(summary[str(f)][key]['mean'])
                stds.append(summary[str(f)][key]['std'])
                valid_fs.append(f)
        if not means:
            continue

        means, stds = np.array(means), np.array(stds)
        x_offset = np.array(valid_fs) * (1 + offsets[i])
        color = cmap[i]

        ax.fill_between(x_offset, means - stds, means + stds, color=color, alpha=0.15)
        ax.errorbar(
            x_offset, means, yerr=stds, fmt='o-', capsize=cap_size,
            elinewidth=1.5, linewidth=line_width, markersize=marker_size,
            color=color, label=key.replace('_',' ').title()
        )

    ax.set_xscale('log', base=2)
    ax.set_xticks([8, 16, 32, 64])
    ax.set_xticklabels(['8', '16', '32', '64'], fontsize=font_tick)
    ax.tick_params(axis='y', labelsize=font_tick)
    ax.grid(True, which='both', linestyle='--', alpha=1.0)
    ax.set_xlabel('Simulation Time Window', fontsize=font_label)
    ax.set_title(title, fontsize=font_label, y=-0.25)
    ax.locator_params(axis='y', nbins=4)
    # Don't add legend here — global legend later


# --------------------------
# Example usage for 3 subplots
# --------------------------
f_samples = [8, 16, 32, 64]
metric_keys = [
    "train_drop_phase2", "train_drop_phase1", "train_drop_phase3",
    "test_drop_phase2", "test_drop_phase1", "test_drop_phase3"
]

ylabel = r"$\delta_{ANN}$"


fig, axes = plt.subplots(1, 3, figsize=(25, 7), sharey=True)

plot_metric_subplot(axes[0], summary_iris, f_samples, metric_keys, ylabel, "(a) Iris")
plot_metric_subplot(axes[1], summary_wine, f_samples, metric_keys, ylabel, "(b) Wine")
plot_metric_subplot(axes[2], summary_bc, f_samples, metric_keys, ylabel, "(a) BC")

axes[0].set_ylabel(ylabel, fontsize=26)

# --- Shared legend ABOVE the subplots ---
handles, labels = axes[0].get_legend_handles_labels()
labels = [
    "train_I", "train_D", "train_ID",
    "test_I", "test_D", "test_ID",
]
fig.legend(handles, labels, loc='upper center', ncol=6, fontsize=26, frameon=True,
           bbox_to_anchor=(0.525, 1.05))  # Push legend above the figure

# Adjust layout to create space above
plt.subplots_adjust(top=0.75, wspace=0.15)
plt.tight_layout(rect=[0, 0, 1, 0.90])  # leave space for legend

plt.savefig("drop.pdf", bbox_inches='tight')
plt.show()

plt.rcParams.update({
    'font.weight': 'bold',
    'axes.labelweight': 'bold',
    'axes.titleweight': 'bold',
    'xtick.labelsize': 26,
    'ytick.labelsize': 26,
    'xtick.major.width': 2,
    'ytick.major.width': 2,
})

def plot_metric_subplot(ax, summary, f_samples, metric_keys, ylabel, title,
                        offset_width=0.05, font_label=26, font_tick=26,
                        line_width=2, marker_size=6, cap_size=4):
    """
    Plot mean ± std for given metrics on a provided Axes (subplot).
    """
    num_keys = len(metric_keys)
    cmap = plt.cm.tab10(np.linspace(0, 1, num_keys))
    offsets = np.linspace(-offset_width, offset_width, num_keys)

    for i, key in enumerate(metric_keys):
        means, stds, valid_fs = [], [], []
        for f in f_samples:
            if key in summary[str(f)]:
                means.append(summary[str(f)][key]['mean'])
                stds.append(summary[str(f)][key]['std'])
                valid_fs.append(f)
        if not means:
            continue

        means, stds = np.array(means), np.array(stds)
        x_offset = np.array(valid_fs) * (1 + offsets[i])
        color = cmap[i]

        ax.fill_between(x_offset, means - stds, means + stds, color=color, alpha=0.15)
        ax.errorbar(
            x_offset, means, yerr=stds, fmt='o-', capsize=cap_size,
            elinewidth=1.5, linewidth=line_width, markersize=marker_size,
            color=color, label=key.replace('_',' ').title()
        )

    ax.set_xscale('log', base=2)
    ax.set_xticks([8, 16, 32, 64])
    ax.set_xticklabels(['8', '16', '32', '64'], fontsize=font_tick)
    ax.tick_params(axis='y', labelsize=font_tick)
    ax.grid(True, which='both', linestyle='--', alpha=1.0)
    ax.set_xlabel('Simulation Time Window', fontsize=font_label)
    ax.set_title(title, fontsize=font_label, y=-0.25)
    ax.locator_params(axis='y', nbins=4)
    # Don't add legend here — global legend later


# --------------------------
# Example usage for 3 subplots
# --------------------------
f_samples = [8, 16, 32, 64]
metric_keys = [
    "train_gain_phase2", "train_gain_phase1", "train_gain_phase3",
    "test_gain_phase2", "test_gain_phase1", "test_gain_phase3"
]
ylabel = r"$\delta_{SNN}$"

fig, axes = plt.subplots(1, 3, figsize=(25, 7), sharey=True)

plot_metric_subplot(axes[0], summary_iris, f_samples, metric_keys, ylabel, "(a) Iris")
plot_metric_subplot(axes[1], summary_wine, f_samples, metric_keys, ylabel, "(b) Wine")
plot_metric_subplot(axes[2], summary_bc, f_samples, metric_keys, ylabel, "(a) BC")

axes[0].set_ylabel(ylabel, fontsize=26)

# --- Shared legend ABOVE the subplots ---
handles, labels = axes[0].get_legend_handles_labels()
labels = [
    "train_I", "train_D", "train_ID",
    "test_I", "test_D", "test_ID",
]
fig.legend(handles, labels, loc='upper center', ncol=6, fontsize=26, frameon=True,
           bbox_to_anchor=(0.525, 1.00))  # Push legend above the figure

# Adjust layout to create space above
plt.subplots_adjust(top=0.75, wspace=0.15)
plt.tight_layout(rect=[0, 0, 1, 0.90])  # leave space for legend

plt.savefig("gain.pdf", bbox_inches='tight')
plt.show()

import matplotlib.pyplot as plt
import numpy as np

f_samples = [8, 16, 32, 64]
phases = ['phase2', 'phase1', 'phase3', 'phase2_', 'phase1_', 'phase3_']

def build_summary(raw_train_test):
    summary = {}
    for method in phases:
        for f in f_samples:
            f_key = str(f)
            if f_key not in summary:
                summary[f_key] = {}

            mean_train = np.array(raw_train_test[f_key][f'mean_spike_train_{method}'])
            mean_test = np.array(raw_train_test[f_key][f'mean_spike_test_{method}'])

            weighted = mean_train * 0.70 + mean_test * 0.30
            summary[f_key][f"spike_count_mean_{method}"] = {
                "mean": np.mean(weighted),
                "std": np.std(weighted)
            }
    return summary

# Build summaries for all datasets
summary_iris = build_summary(raw_iris)
summary_wine = build_summary(raw_wine)
summary_bc = build_summary(raw_bc)

# --------------------------------------------------------------
# Plotting function (bold font + shared legend)
# --------------------------------------------------------------

plt.rcParams.update({
    'font.weight': 'bold',
    'axes.labelweight': 'bold',
    'axes.titleweight': 'bold',
})

def plot_metric_subplot(ax, summary, f_samples, metric_keys, ylabel, title,
                        offset_width=0.10, font_label=26, font_tick=26,
                        line_width=5, marker_size=15, cap_size=10):

    num_keys = len(metric_keys)
    cmap = plt.cm.tab10(np.linspace(0, 1, num_keys))
    offsets = np.linspace(-offset_width, offset_width, num_keys)

    for i, key in enumerate(metric_keys):
        means, stds, valid_fs = [], [], []
        for f in f_samples:
            if key in summary[str(f)]:
                means.append(summary[str(f)][key]["mean"])
                stds.append(summary[str(f)][key]["std"])
                valid_fs.append(f)
        if not means:
            continue

        means, stds = np.array(means), np.array(stds)
        x_offset = np.array(valid_fs) * (1 + offsets[i])
        color = cmap[i]

        ax.fill_between(x_offset, means - stds, means + stds, color=color, alpha=0.30)
        ax.errorbar(x_offset, means, yerr=stds, fmt='o-', capsize=cap_size,
                    elinewidth=5, linewidth=line_width, markersize=marker_size,
                    color=color, label=key.replace('_', ' ').title())

    ax.set_xscale('log', base=2)
    ax.set_xticks([8, 16, 32, 64])
    ax.set_xticklabels(['8', '16', '32', '64'], fontsize=font_tick, fontweight='bold')
    ax.tick_params(axis='y', labelsize=font_tick)
    ax.grid(True, which='both', linestyle='--', alpha=0.4)
    ax.set_xlabel('Simulation Time Window', fontsize=font_label, fontweight='bold')
    ax.set_title(title, fontsize=font_label, y=-0.25, fontweight='bold')
    ax.locator_params(axis='y', nbins=4)

# --------------------------------------------------------------
# Plot three datasets side-by-side
# --------------------------------------------------------------

f_samples = [8, 16, 32, 64]
metric_keys = [f"spike_count_mean_{p}" for p in phases]
ylabel = "Spike"

fig, axes = plt.subplots(1, 3, figsize=(25, 7), sharey=True)

plot_metric_subplot(axes[0], summary_iris, f_samples, metric_keys, ylabel, "(a) Iris")
plot_metric_subplot(axes[1], summary_wine, f_samples, metric_keys, ylabel, "(b) Wine")
plot_metric_subplot(axes[2], summary_bc, f_samples, metric_keys, ylabel, "(c) BC")

axes[0].set_ylabel(ylabel, fontsize=26, fontweight='bold')

from matplotlib import rcParams, font_manager

# Shared legend above
handles, labels = axes[0].get_legend_handles_labels()

labels = ['Spike Count I',
 'Spike Count D',
 'Spike Count ID',
  r'$\delta_{Spike}$ I',
 r'$\delta_{Spike}$ D',
 r'$\delta_{Spike}$ ID',]

fig.legend(
    handles, labels,
    loc='upper center',
    ncol=6,
    frameon=True,
    bbox_to_anchor=(0.52, 1.0),
    prop=font_manager.FontProperties(weight='bold', size=20)  # size + bold together
)

plt.subplots_adjust(top=0.75, wspace=0.15)
plt.tight_layout(rect=[0, 0, 1, 0.90])

plt.savefig("spike.pdf", bbox_inches='tight')
plt.show()
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager

# --------------------------
# USER SETTINGS / INPUTS
# --------------------------
datasets = {
    "wine": "raw_metrics_wine_v2.json",
    "iris": "raw_metrics_iris_v2.json",
    "bc":   "raw_metrics_bc_v2.json"
}
f_samples = [8, 16, 32, 64]

# Make sure these exist in your script/session for spike plotting:
# raw_iris, raw_wine, raw_bc

# Keys / labels (define these exactly as you want)
drop_keys = [
    "train_drop_phase2", "train_drop_phase1", "train_drop_phase3",
    "test_drop_phase2",  "test_drop_phase1",  "test_drop_phase3",
]
gain_keys = [
    "train_gain_phase2", "train_gain_phase1", "train_gain_phase3",
    "test_gain_phase2",  "test_gain_phase1",  "test_gain_phase3",
]
drop_labels = ["train_I", "train_D", "train_ID", "test_I", "test_D", "test_ID"]
gain_labels = ["train_I", "train_D", "train_ID", "test_I", "test_D", "test_ID"]

phases = ['phase2', 'phase1', 'phase3', 'phase2_', 'phase1_', 'phase3_']
spike_keys = [f"spike_count_mean_{p}" for p in phases]
spike_labels = [
    'Spike Count I', 'Spike Count D', 'Spike Count ID',
    r'$\delta_{Spike}$ I', r'$\delta_{Spike}$ D', r'$\delta_{Spike}$ ID'
]

dataset_order = ["iris", "wine", "bc"]
dataset_title = {"iris": "Iris", "wine": "Wine", "bc": "BC"}

# --------------------------
# LOAD DROP/GAIN SUMMARIES
# --------------------------
def build_mean_std_summary_from_json(file_path, f_samples):
    with open(file_path, "r") as f:
        metrics = json.load(f)

    summary = {}
    for fs in f_samples:
        fs_key = str(fs)
        summary[fs_key] = {}
        for key, values in metrics[fs_key].items():
            if key.startswith("confusion"):
                continue
            arr = np.array(values, dtype=float)
            summary[fs_key][key] = {
                "mean": float(np.mean(arr)),
                "std":  float(np.std(arr))
            }
    return summary

summary_drop_gain = {}
for name, fp in datasets.items():
    summary_drop_gain[name] = build_mean_std_summary_from_json(fp, f_samples)

# --------------------------
# BUILD SPIKE SUMMARY
# --------------------------
def build_spike_summary(raw_train_test, f_samples, phases):
    summary = {}
    for method in phases:
        for fs in f_samples:
            fs_key = str(fs)
            summary.setdefault(fs_key, {})

            mean_train = np.array(raw_train_test[fs_key][f"mean_spike_train_{method}"])
            mean_test  = np.array(raw_train_test[fs_key][f"mean_spike_test_{method}"])

            weighted = mean_train * 0.70 + mean_test * 0.30
            summary[fs_key][f"spike_count_mean_{method}"] = {
                "mean": float(np.mean(weighted)),
                "std":  float(np.std(weighted))
            }
    return summary

summary_spike = {
    "iris": build_spike_summary(raw_iris, f_samples, phases),
    "wine": build_spike_summary(raw_wine, f_samples, phases),
    "bc":   build_spike_summary(raw_bc,   f_samples, phases),
}

# --------------------------
# GLOBAL STYLING (BIGGER FONTS)
# --------------------------
plt.rcParams.update({
    'font.weight': 'bold',
    'axes.labelweight': 'bold',
    'axes.titleweight': 'bold',
    'xtick.labelsize': 26,
    'ytick.labelsize': 26,
    'xtick.major.width': 2,
    'ytick.major.width': 2,
})

# --------------------------
# PLOTTING HELPER
# --------------------------
def plot_metric_subplot(ax, summary, f_samples, metric_keys,
                        offset_width=0.05, font_tick=22,
                        line_width=2, marker_size=6, cap_size=4,
                        fill_alpha=0.15):
    num_keys = len(metric_keys)
    cmap = plt.cm.tab10(np.linspace(0, 1, num_keys))
    offsets = np.linspace(-offset_width, offset_width, num_keys)

    for i, key in enumerate(metric_keys):
        means, stds, valid_fs = [], [], []
        for fs in f_samples:
            d = summary.get(str(fs), {})
            if key in d:
                means.append(d[key]["mean"])
                stds.append(d[key]["std"])
                valid_fs.append(fs)
        if not means:
            continue

        means = np.array(means)
        stds  = np.array(stds)
        x_off = np.array(valid_fs) * (1 + offsets[i])
        color = cmap[i]

        ax.fill_between(x_off, means - stds, means + stds, color=color, alpha=fill_alpha)
        ax.errorbar(
            x_off, means, yerr=stds, fmt='o-', capsize=cap_size,
            elinewidth=1.5, linewidth=line_width, markersize=marker_size,
            color=color, label=key
        )

    ax.set_xscale('log', base=2)
    ax.set_xticks([8, 16, 32, 64])
    ax.set_xticklabels(['8', '16', '32', '64'])
    ax.grid(True, which='both', linestyle='--', alpha=0.35)
    ax.tick_params(axis='both', labelsize=font_tick)
    ax.locator_params(axis='y', nbins=4)

# --------------------------
# BUILD COMBINED FIGURE (3x3) with WHITESPACE LEGENDS
# --------------------------
fig, axes = plt.subplots(3, 3, figsize=(26, 15), sharex=True)

for r in range(3):
    for c in range(1, 3):
        axes[r, c].sharey(axes[r, 0])
        axes[r, c].tick_params(labelleft=False)

# Row 1: DROP
for j, ds in enumerate(dataset_order):
    plot_metric_subplot(
        axes[0, j], summary_drop_gain[ds], f_samples, drop_keys,
        offset_width=0.05, line_width=2, marker_size=6, cap_size=4, fill_alpha=0.15
    )

# Row 2: GAIN
for j, ds in enumerate(dataset_order):
    plot_metric_subplot(
        axes[1, j], summary_drop_gain[ds], f_samples, gain_keys,
        offset_width=0.05, line_width=2, marker_size=6, cap_size=4, fill_alpha=0.15
    )

# Row 3: SPIKE
for j, ds in enumerate(dataset_order):
    plot_metric_subplot(
        axes[2, j], summary_spike[ds], f_samples, spike_keys,
        offset_width=0.10, line_width=3, marker_size=8, cap_size=6, fill_alpha=0.30
    )

# Dataset labels inside each subplot
for i in range(3):
    for j, ds in enumerate(dataset_order):
        axes[i, j].text(
            0.03, 0.94, dataset_title[ds],
            transform=axes[i, j].transAxes,
            fontsize=24, fontweight='bold',
            va='top', ha='left'
        )

# Share x-axis: only bottom row shows x tick labels
for i in [0, 1]:
    for j in range(3):
        axes[i, j].tick_params(labelbottom=False)

# Global x label
fig.text(0.2, -0.06, "Simulation Time Window", ha="center",
         fontsize=26, fontweight="bold")

fig.text(0.53, -0.06, "Simulation Time Window", ha="center",
         fontsize=26, fontweight="bold")

fig.text(0.87, -0.06, "Simulation Time Window", ha="center",
         fontsize=26, fontweight="bold")

# Row labels on the left
fig.text(0.02, 0.82, r"$\delta_{ANN}$", rotation=90, va="center",
         fontsize=26, fontweight="bold")
fig.text(0.02, 0.50, r"$\delta_{SNN}$", rotation=90, va="center",
         fontsize=26, fontweight="bold")
fig.text(0.02, 0.19, "Spike", rotation=90, va="center",
         fontsize=26, fontweight="bold")

# IMPORTANT: reserve real whitespace ABOVE the subplot grid
plt.subplots_adjust(left=0.07, right=0.995, top=0.92, bottom=0.00,
                    hspace=0.35, wspace=0.18)

# Legends placed in whitespace bands (figure coords)
h0, _ = axes[0, 0].get_legend_handles_labels()
fig.legend(
    h0, drop_labels, ncol=6, frameon=True,
    loc="upper center",
    bbox_to_anchor=(0.52, 0.97),
    bbox_transform=fig.transFigure,
    prop=font_manager.FontProperties(weight='bold', size=20)
)

h1, _ = axes[1, 0].get_legend_handles_labels()
fig.legend(
    h1, gain_labels, ncol=6, frameon=True,
    loc="upper center",
    bbox_to_anchor=(0.52, 0.64),
    bbox_transform=fig.transFigure,
    prop=font_manager.FontProperties(weight='bold', size=20)
)

h2, _ = axes[2, 0].get_legend_handles_labels()
fig.legend(
    h2, spike_labels, ncol=6, frameon=True,
    loc="upper center",
    bbox_to_anchor=(0.52, 0.315),
    bbox_transform=fig.transFigure,
    prop=font_manager.FontProperties(weight='bold', size=18)
)

# Save one pdf
plt.savefig("all_metrics_compact.pdf", bbox_inches="tight")
plt.show()
```
