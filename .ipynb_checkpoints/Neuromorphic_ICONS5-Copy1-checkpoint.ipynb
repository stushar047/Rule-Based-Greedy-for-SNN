```python
import numpy as np

import matplotlib.pyplot as plt

from sklearn.datasets import load_iris,  load_breast_cancer, load_wine

from sklearn.preprocessing import StandardScaler, MinMaxScaler

from sklearn.model_selection import train_test_split

from mravens_icons import *

import tensorflow as tf

from tensorflow.keras import Sequential

from tensorflow.keras.layers import Dense

import pandas as pd

def one_hot(y,n=3):
    
    y_onehot = np.zeros((len(y),n))

    for i,j in enumerate(y):

        y_onehot[i,j]=1
        
    return y_onehot

data = load_iris()

X,y=data['data'],data['target']

data = load_wine()

X,y=data['data'],data['target']

y_onehot = one_hot(y,np.max(y)+1)

X_norm=(X-np.min(X,axis=0))/(np.max(X,axis=0)-np.min(X,axis=0))

f_sample=32

X_train,X_test,y_train,y_test = train_test_split(X_norm,y_onehot,test_size=0.20,random_state=42, stratify=y)

X_rate_train,X_rate_test = (X_train*f_sample).astype('int'), (X_test*f_sample).astype('int')

# Define the input shape
input_shape = (X_norm.shape[1],)  # Replace input_size with the size of your input

# Create a Sequential model
model = Sequential()

initializer = tf.keras.initializers.HeNormal()

# Add the output layer directly
model.add(Dense(units=y_onehot.shape[1], input_shape=input_shape, activation='sigmoid',use_bias=False, kernel_initializer=initializer))

opt = tf.keras.optimizers.Adam(learning_rate=0.1,beta_1=0.7,beta_2=0.999)

model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])

# Print a summary of the model
model.summary()

# Train the model
history = model.fit(X_train, y_train, epochs=1, batch_size=16, verbose=1, validation_data=(X_test,y_test),shuffle=True)

keras_loss_train, keras_acc_train = model.evaluate(X_train, y_train, verbose=0)

keras_loss_test, keras_acc_test= model.evaluate(X_test, y_test, verbose=0)

keras_weights = model.get_weights()[0]

X_rate_data_train=np.zeros((X_rate_train.shape[0],X_rate_train.shape[1],f_sample))

for m,sample in enumerate(X_rate_train):
    
    for n,feature in enumerate(sample):
        
        X_rate_data_train[m,n,:]=np.zeros(f_sample)
        
        if feature>0:
        
            X_rate_data_train[m,n,0:-1:round(f_sample/feature)]=1

X_rate_data_test=np.zeros((X_rate_test.shape[0],X_rate_test.shape[1],f_sample))

for m,sample in enumerate(X_rate_test):
    
    for n,feature in enumerate(sample):
        
        X_rate_data_test[m,n,:]=np.zeros(f_sample)
        
        if feature>0:
        
            X_rate_data_test[m,n,0:-1:round(f_sample/feature)]=1

def simulate(X_rate_data,keras_weights,f_sample,st_point=[0,1,2,3],n_id =[0,1,2],
                  type_id=["output","output","output"],abs_period=[1,10,1]):

    y_pred = []

    for i in range(len(X_rate_data)):

        net = create_custom_network()

        net.add_stimuli(st_point,n_id,X_rate_data[i],keras_weights)

        for (i,j,k) in zip(n_id,type_id,abs_period):

            net.add_neuron(i,j)
            
            net.neuron[i].abs_period=k

        proc=MRAVENS(-5,5)

        event=process_event(f_sample,net,proc)

        event.apply_spike()

        y_pred.append(np.argmax(np.sum(event.spikes_,axis=0)))
        
    return np.array(y_pred)

ytrain = np.argmax(y_train,axis=1)

ytest = np.argmax(y_test,axis=1)

abs_period = np.array([10,10,10])

acc_train_list, acc_test_list, abs_period_list = [], [], []

for i in range(30):

    y_pred_train = simulate(X_rate_data_train,keras_weights,f_sample,st_point=np.arange(X.shape[1]),n_id =[0,1,2],type_id=["output","output","output"],abs_period=abs_period)

    y_pred_test = simulate(X_rate_data_test,keras_weights,f_sample,st_point=np.arange(X.shape[1]),n_id =[0,1,2],type_id=["output","output","output"],abs_period=abs_period)

    acc_train, acc_test = len(ytrain[y_pred_train==ytrain])/len(ytrain), len(ytest[y_pred_test==ytest])/len(y_test)

    print(f"Iteration:{i}, Keras Accuracy: {keras_acc_train, keras_acc_test} ::: Neuromorphic Accuracy: {acc_train,acc_test}")
    
    spike_dev = np.sum(one_hot(y_pred_train,3)-y_train,axis=0)
    
    print(spike_dev)

    spike_dev_arr = np.zeros_like(spike_dev)
    
    idx = np.argsort(spike_dev)

    spike_dev_arr[idx[0]], spike_dev_arr[idx[-1]]= -1, 0 

    abs_period = (abs_period + spike_dev_arr).astype('int')
    
    acc_train_list.append(acc_train)
    
    acc_test_list.append(acc_test)
    
    abs_period_list.append(abs_period)
    
    print(abs_period)

abs_period = np.array([1,1,1])

acc_train_list, acc_test_list, abs_period_list = [], [], []

for i in range(25):

    y_pred_train = simulate(X_rate_data_train,keras_weights,f_sample,st_point=np.arange(X.shape[1]),n_id =[0,1,2],type_id=["output","output","output"],abs_period=abs_period)

    y_pred_test = simulate(X_rate_data_test,keras_weights,f_sample,st_point=np.arange(X.shape[1]),n_id =[0,1,2],type_id=["output","output","output"],abs_period=abs_period)

    acc_train, acc_test = len(ytrain[y_pred_train==ytrain])/len(ytrain), len(ytest[y_pred_test==ytest])/len(y_test)

    print(f"Iteration:{i}, Keras Accuracy: {keras_acc_train, keras_acc_test} ::: Neuromorphic Accuracy: {acc_train,acc_test}")
    
    spike_dev = np.sum(one_hot(y_pred_train,3)-y_train,axis=0)
    
    print(spike_dev)

    spike_dev_arr = np.zeros_like(spike_dev)
    
    idx = np.argsort(spike_dev)

    spike_dev_arr[idx[0]], spike_dev_arr[idx[1]]= 0, 1 

    abs_period = (abs_period + spike_dev_arr).astype('int')
    
    acc_train_list.append(acc_train)
    
    acc_test_list.append(acc_test)
    
    abs_period_list.append(abs_period)
    
    print(abs_period)

sublist = np.array(abs_period_list)

xticks_labels =[]

for item in sublist:
    
    xticks_labels.append(f"{item[0]},{item[1]},{item[2]}")

plt.figure(figsize=(20,6))

plt.rcParams.update({'font.size': 14})

plt.plot(acc_test_list[:30], linewidth = 4)

plt.plot(acc_train_list[:30],linewidth = 4)

plt.xticks(range(30), xticks_labels[:30])

plt.legend(["train_accuracy","test_accuracy"])

plt.xlabel("Absoluate Period: (Neuron 1, Neuron 2, Neuron 3)")

plt.ylabel("Accuracy")

plt.title(f"Dataset(Wine), Software accuracy: Train:{keras_acc_train}, Test:{keras_acc_test}")

# plt.savefig("Wine_abs.jpg",dpi=600,bbox_inches='tight')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, load_wine
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from mravens_icons import *
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd


# -------------------------------
# Utility Functions
# -------------------------------
def one_hot(y, n=3):
    y_onehot = np.zeros((len(y), n))
    for i, j in enumerate(y):
        y_onehot[i, j] = 1
    return y_onehot


# -------------------------------
# Load Dataset (Wine)
# -------------------------------
data = load_wine()
X, y = data['data'], data['target']

# One-hot encode targets
y_onehot = one_hot(y, np.max(y) + 1)

# Normalize data between 0 and 1
X_norm = (X - np.min(X, axis=0)) / (np.max(X, axis=0) - np.min(X, axis=0))

# Quantize into spike rate levels
f_sample = 32
X_train, X_test, y_train, y_test = train_test_split(
    X_norm, y_onehot, test_size=0.20, random_state=42, stratify=y
)
X_rate_train = (X_train * f_sample).astype('int')
X_rate_test = (X_test * f_sample).astype('int')


# -------------------------------
# Define and Train Keras Model
# -------------------------------
input_shape = (X_norm.shape[1],)
model = Sequential()
initializer = tf.keras.initializers.HeNormal()

model.add(Dense(
    units=y_onehot.shape[1],
    input_shape=input_shape,
    activation='sigmoid',
    use_bias=False,
    kernel_initializer=initializer
))

opt = tf.keras.optimizers.Adam(learning_rate=0.1, beta_1=0.7, beta_2=0.999)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
model.summary()

history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=16,
    verbose=1,
    validation_data=(X_test, y_test),
    shuffle=True
)

keras_loss_train, keras_acc_train = model.evaluate(X_train, y_train, verbose=0)
keras_loss_test, keras_acc_test = model.evaluate(X_test, y_test, verbose=0)
keras_weights = model.get_weights()[0]


# -------------------------------
# Spike Encoding
# -------------------------------
X_rate_data_train = np.zeros((X_rate_train.shape[0], X_rate_train.shape[1], f_sample))
for m, sample in enumerate(X_rate_train):
    for n, feature in enumerate(sample):
        if feature > 0:
            X_rate_data_train[m, n, 0:-1:round(f_sample / feature)] = 1

X_rate_data_test = np.zeros((X_rate_test.shape[0], X_rate_test.shape[1], f_sample))
for m, sample in enumerate(X_rate_test):
    for n, feature in enumerate(sample):
        if feature > 0:
            X_rate_data_test[m, n, 0:-1:round(f_sample / feature)] = 1


# -------------------------------
# Simulation Function
# -------------------------------
def simulate(X_rate_data, keras_weights, f_sample, st_point=[0, 1, 2, 3],
             n_id=[0, 1, 2], type_id=["output", "output", "output"], abs_period=[1, 10, 1]):

    y_pred = []
    for i in range(len(X_rate_data)):
        net = create_custom_network()
        net.add_stimuli(st_point, n_id, X_rate_data[i], keras_weights)

        for (nid, ttype, period) in zip(n_id, type_id, abs_period):
            net.add_neuron(nid, ttype)
            net.neuron[nid].abs_period = period

        proc = MRAVENS(-5, 5)
        event = process_event(f_sample, net, proc)
        event.apply_spike()

        y_pred.append(np.argmax(np.sum(event.spikes_, axis=0)))
    return np.array(y_pred)

"""
-------------------------------
# Neuromorphic Training Loop
-------------------------------
"""
ytrain = np.argmax(y_train, axis=1)
ytest = np.argmax(y_test, axis=1)

abs_period = np.array([10, 10, 10])
acc_train_list, acc_test_list, abs_period_list = [], [], []

for i in range(30):
    y_pred_train = simulate(
        X_rate_data_train, keras_weights, f_sample,
        st_point=np.arange(X.shape[1]),
        n_id=[0, 1, 2],
        type_id=["output", "output", "output"],
        abs_period=abs_period
    )

    y_pred_test = simulate(
        X_rate_data_test, keras_weights, f_sample,
        st_point=np.arange(X.shape[1]),
        n_id=[0, 1, 2],
        type_id=["output", "output", "output"],
        abs_period=abs_period
    )

    acc_train = np.mean(y_pred_train == ytrain)
    acc_test = np.mean(y_pred_test == ytest)
    print(f"Iteration: {i}, Keras Accuracy: ({keras_acc_train:.3f}, {keras_acc_test:.3f}) ::: Neuromorphic Accuracy: ({acc_train:.3f}, {acc_test:.3f})")

    spike_dev = np.sum(one_hot(y_pred_train, 3) - y_train, axis=0)
    print(spike_dev)

    spike_dev_arr = np.zeros_like(spike_dev)
    idx = np.argsort(spike_dev)
    spike_dev_arr[idx[0]], spike_dev_arr[idx[-1]] = -1, 0

    abs_period = (abs_period + spike_dev_arr).astype('int')

    acc_train_list.append(acc_train)
    acc_test_list.append(acc_test)
    abs_period_list.append(abs_period)
    print(abs_period)


# -------------------------------
# Second Neuromorphic Phase
# -------------------------------
abs_period = np.array([1, 1, 1])
acc_train_list, acc_test_list, abs_period_list = [], [], []

for i in range(25):
    y_pred_train = simulate(
        X_rate_data_train, keras_weights, f_sample,
        st_point=np.arange(X.shape[1]),
        n_id=[0, 1, 2],
        type_id=["output", "output", "output"],
        abs_period=abs_period
    )

    y_pred_test = simulate(
        X_rate_data_test, keras_weights, f_sample,
        st_point=np.arange(X.shape[1]),
        n_id=[0, 1, 2],
        type_id=["output", "output", "output"],
        abs_period=abs_period
    )

    acc_train = np.mean(y_pred_train == ytrain)
    acc_test = np.mean(y_pred_test == ytest)
    print(f"Iteration: {i}, Keras Accuracy: ({keras_acc_train:.3f}, {keras_acc_test:.3f}) ::: Neuromorphic Accuracy: ({acc_train:.3f}, {acc_test:.3f})")

    spike_dev = np.sum(one_hot(y_pred_train, 3) - y_train, axis=0)
    print(spike_dev)

    spike_dev_arr = np.zeros_like(spike_dev)
    idx = np.argsort(spike_dev)
    spike_dev_arr[idx[0]], spike_dev_arr[idx[-1]] = 0, 1

    abs_period = (abs_period + spike_dev_arr).astype('int')

    acc_train_list.append(acc_train)
    acc_test_list.append(acc_test)
    abs_period_list.append(abs_period)
    print(abs_period)

# -------------------------------
# Plotting Results
# -------------------------------
# -------------------------------
# Plotting Results (Fixed)
# -------------------------------
sublist = np.array(abs_period_list)
xticks_labels = [f"{a},{b},{c}" for (a, b, c) in sublist]

num_points = len(acc_test_list)  # adaptively match number of iterations
plt.figure(figsize=(20, 6))
plt.rcParams.update({'font.size': 14})

plt.plot(acc_test_list, linewidth=4)
plt.plot(acc_train_list, linewidth=4)

# ✅ dynamically set correct tick positions and labels
plt.xticks(range(num_points), xticks_labels, rotation=45)

plt.legend(["train_accuracy", "test_accuracy"])
plt.xlabel("Absolute Period: (Neuron 1, Neuron 2, Neuron 3)")
plt.ylabel("Accuracy")
plt.title(f"Dataset (Wine), Software Accuracy: Train={keras_acc_train:.3f}, Test={keras_acc_test:.3f}")
plt.tight_layout()
plt.show()
```
